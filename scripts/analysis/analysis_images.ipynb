### **Preparando dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
file_path = "C:\\Users\\isaja\\Documents\\dcc\\e\\crisis\\data\\crisis_images_dataset.csv"
df = pd.read_csv(file_path)
df.head()
```

### **Muestra aleatoria**

```python
import random

random_state = random.randint(0,1000)

# Tomar 6 imágenes aleatorias
sample_images = df['image_path'].sample(6, random_state=random_state)

default_path = "C:/Users/isaja/Documents/dcc/e/CrisisMMD_v2.0/CrisisMMD_v2.0/"

plt.figure(figsize=(12, 8))

for i, image in enumerate(sample_images):
    real_path = default_path + image
    img = mpimg.imread(real_path)
    
    # 2 filas, 3 columnas
    plt.subplot(2, 3, i + 1)
    plt.imshow(img)
    plt.axis('off')

plt.show()
```

### **Balance de informativo vs no informativo**

```python
# Cantidad de informativos vs no informativos

print("Balance de imágenes informativas vs no informativas:\n")

informativos = df[df['image_info'] == 'informative'].shape[0]
no_informativos = df[df['image_info'] == 'not_informative'].shape[0]
totales = informativos + no_informativos

print(f"Total de imágenes: {totales}\n")

print(f"Cantidad de imágenes informativas: {informativos}")
print(f"Cantidad de imágenes no informativas: {no_informativos}\n")

porcentaje_informativos = (informativos / totales) * 100
porcentaje_no_informativos = (no_informativos / totales) * 100
print(f"Porcentaje de imágenes informativas: {porcentaje_informativos:.2f}%")
print(f"Porcentaje de imágenes no informativas: {porcentaje_no_informativos:.2f}%")
```
